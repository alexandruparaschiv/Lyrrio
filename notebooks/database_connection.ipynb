```python
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
from sqlalchemy import MetaData, Table, select
```

```python
engine = create_engine('postgresql://postgres:psql123@localhost:5432/songsdb')
Session = sessionmaker(bind=engine)
session = Session()
```

```python
session.query()
```

```python
print(engine.table_names())
```

```python
metadata = MetaData()
songs = Table('songs',metadata,autoload=True,autoload_with=engine)
```

```python
print(repr(songs))
```

```python
result = engine.execute('SELECT * FROM songs')

rows = result.fetchall()
print(rows)        # [(3, 'Melinda', 100), (4, 'George', 100)]

result.close()
```

```python
ins = songs.insert().values(
       id=3,
      author='Hello',
      title='World',
      lyrics='lyrics')
conn = engine.connect()
conn.execute(ins)
```
